# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, ne_chunk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# Function to scrape IMDB movie reviews
def scrape_imdb_reviews(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    reviews = []
    for review in soup.find_all('div', class_='text show-more__control'):
        review_text = review.text.strip()
        reviews.append(review_text)
    return reviews

# IMDB movie "Inception" user reviews URL
imdb_url = 'https://www.imdb.com/title/tt1375666/reviews'

# Scrape IMDB movie reviews
reviews = scrape_imdb_reviews(imdb_url)

# Convert to DataFrame and save to CSV
df = pd.DataFrame(reviews, columns=['review_text'])
df.to_csv('imdb_reviews.csv', index=False)

```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Function to clean text data: Remove noise (special characters and punctuations)
def remove_noise(text):
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

# Apply noise removal function to the review_text column
df['cleaned_text_noise'] = df['review_text'].apply(remove_noise)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Noise Removal):")
    print(row['cleaned_text_noise'])
    print("="*50)
```

```python
# Function to clean text data: Remove numbers
def remove_numbers(text):
    cleaned_text = re.sub(r'\d+', '', text)
    return cleaned_text

# Apply number removal function to the review_text column
df['cleaned_text_numbers'] = df['cleaned_text_noise'].apply(remove_numbers)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Numbers Removal):")
    print(row['cleaned_text_numbers'])
    print("="*50)
```

```python
# Function to clean text data: Remove stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    cleaned_tokens = [word for word in tokens if word.lower() not in stop_words]
    cleaned_text = ' '.join(cleaned_tokens)
    return cleaned_text

# Apply stopwords removal function to the review_text column
df['cleaned_text_stopwords'] = df['cleaned_text_numbers'].apply(remove_stopwords)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Stopwords Removal):")
    print(row['cleaned_text_stopwords'])
    print("="*50)
```

```python
# Function to clean text data: Lowercase all texts
def lowercase_text(text):
    cleaned_text = text.lower()
    return cleaned_text

# Apply lowercase function to the review_text column
df['cleaned_text_lowercase'] = df['cleaned_text_stopwords'].apply(lowercase_text)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Lowercasing):")
    print(row['cleaned_text_lowercase'])
    print("="*50)
```

```python
# Function to clean text data: Stemming
def stemming_text(text):
    stemmer = PorterStemmer()
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    cleaned_text = ' '.join(stemmed_tokens)
    return cleaned_text

# Apply stemming function to the review_text column
df['cleaned_text_stemming'] = df['cleaned_text_lowercase'].apply(stemming_text)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Stemming):")
    print(row['cleaned_text_stemming'])
    print("="*50)
```

```python
# Function to clean text data: Lemmatization
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    cleaned_text = ' '.join(lemmatized_tokens)
    return cleaned_text

# Apply lemmatization function to the review_text column
df['cleaned_text_lemmatization'] = df['cleaned_text_stemming'].apply(lemmatize_text)

# Display original and cleaned text for comparison
for index, row in df.head().iterrows():
    print("Original Text:")
    print(row['review_text'])
    print("Cleaned Text (Lemmatization):")
    print(row['cleaned_text_lemmatization'])
    print("="*50)
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Function to conduct syntax and structure analysis
def analyze_text(text):
    # Parts of Speech (POS) Tagging
    pos_tags = pos_tag(word_tokenize(text))
    # Calculate counts of Noun, Verb, Adjective, Adverb
    noun_count = len([word for word, tag in pos_tags if tag.startswith('NN')])
    verb_count = len([word for word, tag in pos_tags if tag.startswith('VB')])
    adj_count = len([word for word, tag in pos_tags if tag.startswith('JJ')])
    adv_count = len([word for word, tag in pos_tags if tag.startswith('RB')])

    # Constituency Parsing and Dependency Parsing (not implemented here)

    # Named Entity Recognition
    named_entities = ne_chunk(pos_tags)
    named_entities_dict = {}
    for entity in named_entities:
        if isinstance(entity, nltk.tree.Tree):
            entity_text = ' '.join([word for word, tag in entity.leaves()])
            entity_type = entity.label()
            if entity_type not in named_entities_dict:
                named_entities_dict[entity_type] = []
            named_entities_dict[entity_type].append(entity_text)

    return noun_count, verb_count, adj_count, adv_count, named_entities_dict

# Analyze text and print results
for index, row in df.iterrows():
    print(f"Review {index + 1}:")
    print("Original Text:", row['review_text'])
    noun_count, verb_count, adj_count, adv_count, named_entities = analyze_text(row['cleaned_text'])
    print("Noun Count:", noun_count)
    print("Verb Count:", verb_count)
    print("Adjective Count:", adj_count)
    print("Adverb Count:", adv_count)
    print("Named Entities:", named_entities)
    print("="*50)

```
